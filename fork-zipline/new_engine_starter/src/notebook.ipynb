```python
import pandas as pd

# 假设 df 有 symbol, date, close 字段
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['symbol', 'date'])

# 计算7日动量（收盘价与7天前的比值）
df['momentum_7'] = df.groupby('symbol')['close'].pct_change(periods=7)

# 判断是否连续7天增长
def is_7_days_up(prices):
    return (prices.diff() > 0).rolling(7).sum() == 7

df['is_7_up'] = df.groupby('symbol')['close'].apply(is_7_days_up).reset_index(level=0, drop=True)

# 筛选出满足条件的股票
selected = df[df['is_7_up'] == True]['symbol'].unique()
print("连续7天增长的股票：", selected)
```
